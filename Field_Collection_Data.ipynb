```python
#read original form data
form_1 = pd.read_csv("ID_SCREEN.csv")
mb = pd.read_csv("MB_+VE.csv")
final_template = pd.read_csv("ImportTemplate2.csv")
#mb_data = pd.read_csv("Final_Template_Data.csv")

#loop through forms provided
# for each form containing collected data
for file in range (4):
    print("\nOpening file"+ " " + str(file))
    #store form in variable
    orig_form = pd.read_csv("ID_SCREEN" + str(file) + ".csv")
    
    #store values with MB+ve result(Identification_and_Screening csv)
    pos_result = orig_form.loc[orig_form['MB +ive'] == 'POSITIVE']
    #rename identification column
    pos_result = pos_result.rename(columns = {'Sample ID': 'Sample_ID'})
    
    # ID FORMAT IS CHANGED TO MATCH MB+VE FILE
    #match and arrangeID formats
    split_id = mb["Sample ID"].str.split('/', expand = True)
    split_id[2] = '09'
    split_id_cols = [0,1,2,3]
    split_id['Sample_ID'] = split_id[split_id_cols].apply(lambda row: '/'.join(row.values.astype(str)), axis=1)
    arranged_id = split_id['Sample_ID']
    
    #read MB+ve file and merge column with modified ID column
    final_mb = mb
    final_mb = final_mb.join(arranged_id)
    
    #drop duplicate column and obtain final screening results file
    final_mb.drop(['Sample ID'], axis = 'columns', inplace = True) 
    final_mb = final_mb[['Sample_ID','Target ct','Efficiency','Target R2','Host ct','Efficiency2','Host R2']]
    
    #merge with original MB+ve file to find common columns if any
    final_mb = final_mb.merge(pos_result, how = "right")
    
    #write data to template file
    #store columns in variables
    mb_id = final_mb.loc[:,'Sample_ID']
    mb_tar = final_mb.loc[:,'Target ct']
    mb_eff = final_mb.loc[:,'Efficiency']
    mb_tar_r2 = final_mb.loc[:,'Target R2']
    mb_host = final_mb.loc[:,'Host ct']
    mb_eff2 = final_mb.loc[:,'Efficiency2']
    mb_host_r2 = final_mb.loc[:,'Host R2']
    mb_egg = final_mb.loc[:,'Egg count']
    mb_hatch = final_mb.loc[:,'Hatching']

    #write variables to template columns
    final_template.loc[:,'id'] = mb_id
    final_template.loc[:,'mb_ct'] = mb_tar
    final_template.loc[:,'mb_efficiency'] = mb_eff
    final_template.loc[:,'r2_target'] = mb_tar_r2
    final_template.loc[:,'ct_host'] = mb_host
    final_template.loc[:,'host_efficiency'] = mb_eff2
    final_template.loc[:,'r2_host'] = mb_host_r2
    final_template.loc[:,'eggs'] = mb_egg
    final_template.loc[:,'hatch'] = mb_hatch

#save final results data    
final_mb.to_csv("Final_Template_Data.csv", index=False)
#drop unwanted column
final_template.drop(['Unnamed: 22'], axis = 'columns', inplace = True)
#write to final csv file
final_template.to_csv("ImportTemplate2_updated.csv", index=False)
#display files
display(final_mb)
display(final_template)
```
